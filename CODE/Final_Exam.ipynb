```r
data(mtcars)

mpg<-mtcars$mpg
wt<-mtcars$wt
hp<-mtcars$hp

cat("mpg\n")
cat("Mean:", mean(mpg), "\n")
cat("SD:", sd(mpg), "\n")
cat("Median:", median(mpg), "\n")
cat("IQR:", IQR(mpg), "\n\n")

cat("wt\n")
cat("Mean:", mean(wt), "\n")
cat("SD:", sd(wt), "\n")
cat("Median:", median(wt), "\n")
cat("IQR:", IQR(wt), "\n\n")

cat("hp\n")
cat("Mean:", mean(hp), "\n")
cat("SD:", sd(hp), "\n")
cat("Median:", median(hp), "\n")
cat("IQR:", IQR(hp), "\n")
```

```r
data(mtcars)

mpg<-mtcars$mpg
wt<-mtcars$wt
hp<-mtcars$hp

summary_table<-data.frame(
  variable=c("mpg","wt","hp"),
  mean=c(mean(mpg), mean(wt), mean(hp)),
  sd=c(sd(mpg), sd(wt), sd(hp)),
  median=c(median(mpg), median(wt), median(hp)),
  IQR=c(IQR(mpg), IQR(wt), IQR(hp))
)

summary_table
```

```r
data(mtcars)

mpg<-mtcars$mpg

set.seed(1)

n<-5
B<-10000

mean_mpg<-numeric(B)

for(i in 1:B){
  s<-sample(mpg, size=n, replace=TRUE)
  mean_mpg[i]<-mean(s)
}

mean(mean_mpg)
sd(mean_mpg)
```

```r
hist(mean_mpg,
     main="Sampling distribution of mean(mpg)",
     xlab="sample mean of mpg",
     breaks=30)
```

```r
mean(mean_mpg)
sd(mean_mpg)

mean(mpg)
sd(mpg)/sqrt(n)

```
